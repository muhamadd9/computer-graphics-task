```python
import matplotlib.pyplot as plt
import numpy as np
import math

plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
plt.rc('figure', figsize=(26, 8))
```

# Translation


```python

x = [1, 3, 6]
y = [0, 6, 0]

tx = 2
ty = 5

translated_x = [vertex + tx for vertex in x]
translated_y = [vertex + ty for vertex in y]

plt.plot(x + [x[0]], y + [y[0]], color='#a45', linestyle='-', marker='o', label='Original')

plt.plot(translated_x + [translated_x[0]], translated_y + [translated_y[0]], color='#54cccc', linestyle='--', marker='x', label='Translated')


plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Translation of Triangle')
plt.show()
```

# Rotation


```python


x = [0, 3]
y = [0, 3]

angle = 45

angle_rad = math.radians(angle)

rotated_x = [vertex * math.cos(angle_rad) - y[i] * math.sin(angle_rad) for i, vertex in enumerate(x)]
rotated_y = [vertex * math.sin(angle_rad) + y[i] * math.cos(angle_rad) for i, vertex in enumerate(x)]

plt.plot(x, y, color='#a45', linestyle='-', marker='o', label='Original')

plt.plot(rotated_x, rotated_y, color='#54cccc', linestyle='--', marker='x', label='Rotated')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

# Scaling


```python

x = [0, 4, 4, 0, 0]
y = [0, 0, 3, 3, 0]

sx = 2
sy = 4

scaled_x = [vertex * sx for vertex in x]
scaled_y = [vertex * sy for vertex in y]

plt.plot(x, y,  color='#a45', linestyle='-', marker='o', label='Original')

plt.plot(scaled_x, scaled_y, color='#54cccc', linestyle='--', marker='x', label='Scaled')



plt.legend()
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

# Reflection


```python
import matplotlib.pyplot as plt

x = [0, 2, 3, 0]
y = [0, 0, 1, 0]

sx = -1
sy = 1

reflected_x = [vertex * sx for vertex in x]
reflected_y = [vertex * sy for vertex in y]

plt.plot(x, y,  color='#a45', linestyle='-', marker='o', label='Original')

plt.plot(reflected_x, reflected_y, color='#54cccc', linestyle='--', marker='x', label='Reflected')



plt.legend()
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

# Shearing


```python
x = [0, 4, 4, 0, 0]
y = [0, 0, 2, 2, 0]

sx = 2
sy = 4

sheared_x = [vertex + sy * y[i] for i, vertex in enumerate(x)]
sheared_y = [vertex + sx * x[i] for i, vertex in enumerate(y)]

plt.plot(x, y,  color='#a45', linestyle='-', marker='o', label='Original')

plt.plot(sheared_x, sheared_y, color='#54cccc', linestyle='--', marker='x', label='Sheared')


plt.legend()
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

# circle drawing


```python
import matplotlib.pyplot as plt

def bresenham_circle(x0, y0, radius):
    x = radius
    y = 0
    p = 3 - 2 * radius

    points = []

    while x >= y:
        points.append((x0 + x, y0 + y))
        points.append((x0 - x, y0 + y))
        points.append((x0 + x, y0 - y))
        points.append((x0 - x, y0 - y))
        points.append((x0 + y, y0 + x))
        points.append((x0 - y, y0 + x))
        points.append((x0 + y, y0 - x))
        points.append((x0 - y, y0 - x))

        if p <= 0:
            y += 1
            p += 4 * y + 6
        else:
            x -= 1
            y += 1
            p += 4 * (y - x) + 10

    return points

points = bresenham_circle(50, 50, 30)

plt.scatter([x for x, y in points], [y for x, y in points], color='#a45')
circle = plt.Circle((50, 50), 30, fill=False)
plt.gcf().gca().add_artist(circle)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bresenham Circle')
plt.grid(True)
plt.show()
```

# line drawing


```python
import matplotlib.pyplot as plt

def bresenham_line(x0, y0, x1, y1):
    points_line = []
    dx = abs(x1 - x0)
    dy = abs(y1 - y0)
    p = 2 * dy - dx
    x = x0
    y = y0

    x_step = 1 if x0 < x1 else -1
    y_step = 1 if y0 < y1 else -1

    points_line.append((x, y))

    while x != x1:
        x += x_step
        if p < 0:
            p += 2 * dy
        else:
            y += y_step
            p += 2 * (dy - dx)
        points_line.append((x, y))

    return points_line

# Example 
line_points = bresenham_line(1, 1, 8, 5)

x_line, y_line = zip(*line_points)

plt.plot(x_line, y_line, marker='o', label='Bresenham Line')  
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bresenham Line')
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box') 
plt.show()
```

# Cubic curve


```python

def cubic_curve(a, b, c, d, xlim=(-10, 10), ylim=(-10, 10), num_points=1000):
    x = np.linspace(xlim[0], xlim[1], num_points)
    y = a * x**3 + b * x**2 + c * x + d

    plt.plot(x, y)  
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.show()

a = 1
b = -2
c = -1
d = 2
cubic_curve(a, b, c, d)
```

# Explicit Curve


```python
def explicit_curve(function, xlim=(-10, 10), ylim=(-10, 10), num_points=1000):
    x = np.linspace(xlim[0], xlim[1], num_points)
    y = eval(function)  

    plt.plot(x, y) 
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.show()


function = 'np.sin(x)'
explicit_curve(function)
```

# Hermite curve


```python
def hermite_curve(control_points, tangent_vectors, num_points=100):
    t = np.linspace(0, 1, num_points)
    n = len(control_points)
    curve = np.zeros((num_points, 2))

    for i in range(n - 1):
        p0, p1 = control_points[i], control_points[i + 1]
        t0, t1 = tangent_vectors[i], tangent_vectors[i + 1]
        h00 = 2 * t**3 - 3 * t**2 + 1
        h10 = t**3 - 2 * t**2 + t
        h01 = -2 * t**3 + 3 * t**2
        h11 = t**3 - t**2

        x = h00 * p0[0] + h10 * t0[0] + h01 * p1[0] + h11 * t1[0]
        y = h00 * p0[1] + h10 * t0[1] + h01 * p1[1] + h11 * t1[1]

        curve += np.column_stack((x, y))

    plt.plot(curve[:, 0], curve[:, 1]) 
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.show()


control_points = [(0, 0), (1, 2), (3, -1), (4, 3)]  # Control
tangent_vectors = [(1, 1), (2, -1), (-1, -2), (3, 1)]  # Tangent
hermite_curve(control_points, tangent_vectors)
```

# implicit curve


```python
def implicit_curve(equation, xlim=(-10, 10), ylim=(-10, 10), num_points=1000):
    x = np.linspace(xlim[0], xlim[1], num_points)
    y = np.linspace(ylim[0], ylim[1], num_points)
    X, Y = np.meshgrid(x, y)

    Z = eval(equation)  

    plt.contour(X, Y, Z, [0], colors='#a45')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.show()
    
equation = 'X**2 + Y**2 - 16'  
implicit_curve(equation)
```

# Quadratic Curve

```python
def quadratic_curve(a, b, c, xlim=(-10, 10), ylim=(-10, 10), num_points=1000):
    x = np.linspace(xlim[0], xlim[1], num_points)
    y = a * x**2 + b * x + c

    plt.plot(x, y)  
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.show()


a = 1
b = 2
c = 1
quadratic_curve(a, b, c)
```

```python
from scipy.interpolate import splprep, splev

def spline_curve(x, y, num_points=1000):
    tck, u = splprep([x, y], s=0)
    u_new = np.linspace(u.min(), u.max(), num_points)
    x_new, y_new = splev(u_new, tck)

    plt.plot(x_new, y_new)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.show()

x = [0, 1, 3, 4]  
y = [0, 2, -1, 3]  
spline_curve(x, y)
```
